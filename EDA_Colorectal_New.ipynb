```python
#Comparison of Gender, Stage of Cancer, Smoking Status, and Cancer Diagnosis
plt.subplot(221)

df['Sex'].value_counts().plot(kind='bar', title='Gender of patients', figsize=(16,9))

plt.xticks(rotation=0)

plt.subplot(222)

df['cancer_staging'].value_counts().plot(kind='bar', title='Cancer Stage')

plt.xticks(rotation=0)

plt.subplot(223)

df['_Ismoking4_2'].value_counts().plot(kind='bar', title='Were Smokers')

plt.xticks(rotation=0)

plt.subplot(224)

df['Case_Control'].value_counts().plot(kind='bar', title='Were Cancerous')

plt.xticks(rotation=0)

plt.show()
```

```python
#Data distribution based on diagnosis vs. gender
sns.set(rc={"figure.figsize":(16, 9)})
sns.distplot(df[['Case_Control', 'Sex']])
```

```python
#Selecting Some Features (All Features is too big for the runtime)
df_select = df[['Case_Control', 'Sex', 'Nausea', 'Abdominal pain/cramps general', 'Vomiting', 'Diarrhoea', 'Rectal bleeding', 'Change faeces/bowel movements', 'Elevated blood pressure']]
```

```python
#Correlation between symptoms
corr = df_select.corr()
sns.heatmap(corr, annot=True, square=True)
plt.yticks(rotation=0)
plt.show()
```

```python
#Symptoms by Gender
df.groupby('Sex')[['Case_Control', 'Nausea', 'Vomiting', 'Diarrhoea', 'Rectal bleeding', 'Change faeces/bowel movements', 'Elevated blood pressure' ]].mean().T.plot(figsize=(12,8))
```

```python
#Symptoms by Gender
df.groupby('Sex')[['Case_Control', 'Nausea', 'Vomiting', 'Diarrhoea', 'Rectal bleeding', 'Change faeces/bowel movements', 'Elevated blood pressure' ]].mean().T.plot(kind='barh', figsize=(12,8))
```

```python
#Symptom Counts
sns.set(rc={"figure.figsize":(16, 16)})
sns.displot(df_select, bins=2)
```

```python
g = sns.PairGrid(df_select)
g.map(sns.barplot)
```

```python
g = sns.barplot(data=df_select)
g.set_xticklabels(
    labels=["Case_Control", "Sex", "Nausea", "Abdominal pain/cramps", "Vomiting", "Diarrhoea", "Rectal bleeding", "Change faeces/bowel movements", "Elevated blood pressure" ], rotation=30)
# Show the plot
plt.show()
```

```python
#distribution vs. density
g = sns.violinplot(data=df_select)
g.set_xticklabels(
    labels=["Case_Control", "Sex", "Nausea", "Abdominal pain/cramps", "Vomiting", "Diarrhoea", "Rectal bleeding", "Change faeces/bowel movements", "Elevated blood pressure" ], rotation=30)
# Show the plot
plt.show()
```

```python
#Kernal Density
g = sns.kdeplot(data=df_select)
g.set_xticklabels(
    labels=["Case_Control", "Sex", "Nausea", "Abdominal pain/cramps", "Vomiting", "Diarrhoea", "Rectal bleeding", "Change faeces/bowel movements", "Elevated blood pressure" ], rotation=30)
# Show the plot
plt.show()
```
